# LGBM


##LGBM Default

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_values_subset, train_labels, test_size=0.3, random_state=45)
```

```python
import lightgbm as lgbm

lgb_d = lgbm.LGBMClassifier()
lgb_d.fit(X_train,y_train.values.ravel())
```

```python
from sklearn.metrics import f1_score

in_sample_preds_d = lgb_d.predict(X_test)
f1_score(y_test, in_sample_preds_d, average='micro')
```

```python
sorted_idx = lgb_d.feature_importances_.argsort()
from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))
plt.barh(X_train.columns[sorted_idx][62:82], lgb_d.feature_importances_[sorted_idx][62:82])
#plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Importancia')
plt.title('Importancia Features con LGBM con RS')
plt.grid(True)
plt.show()
```

##Busqueda hiper parametros 2

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_values_subset, train_labels, test_size=0.3, random_state=45)
```

```python
import lightgbm as lgbm
from sklearn.model_selection import train_test_split, RandomizedSearchCV
params = {'colsample_bytree': [0.5,1.0,2.0],
          'learning_rate': [0.5,1.0,2.0],
          'max_depth': [-1, 5, 10],
          'min_child_samples':[15,20,25],
          'min_child_weight':[0.001,0.2,0.5],
          'subsample_for_bin':[150000,200000,250000],
          'iterations': [500],}
lightgb = lgbm.LGBMClassifier()
lightgb_model = RandomizedSearchCV(lightgb, params, scoring="f1_micro", cv=5)
lightgb_model.fit(X_train, y_train)
```

```python
lightgb_model.best_score_
```

```python
lightgb_model.best_params_
```

```python
import lightgbm as lgbm

#Entreno con hiperparametros obtenidos
lgb = lgbm.LGBMClassifier(n_jobs=-1,colsample_bytree=1.0,learning_rate=1.0,max_depth=5,
                          min_child_samples=20,min_child_weight=0.5,subsample_for_bin=200000,
                          loss_fuction='MultiClass')
lgb.fit(X_train,y_train.values.ravel())
```

```python
from sklearn.metrics import f1_score

in_sample_preds = lgb.predict(X_test)
f1_score(y_test, in_sample_preds, average='micro')
```

```python
sorted_idx = lgb.feature_importances_.argsort()
from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))
plt.barh(X_train.columns[sorted_idx][62:82], lgb.feature_importances_[sorted_idx][62:82])
#plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Importancia')
plt.title('Importancia Features con LGBM con RS')
plt.grid(True)
plt.show()
```
