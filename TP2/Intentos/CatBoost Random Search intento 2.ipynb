```python
import catboost as cb
from sklearn.model_selection import train_test_split, RandomizedSearchCV
params = {'depth': np.arange(5, 15, 1),
          'learning_rate': np.arange(0.15, 0.25, 0.01),
          'l2_leaf_reg': np.arange(9, 16, 1),
          'iterations': np.arange(400,600,20)}
cb = cb.CatBoostClassifier()
cb_model = RandomizedSearchCV(cb, params, scoring="f1_micro", cv=3)
cb_model.fit(X_train, y_train)
```

```python
categorical_features_indices = np.where(X_train.dtypes =='object')[0]
```

```python
#Entreno con los mejores parametros que me dio el grid search
import catboost as cb
clf = cb.CatBoostClassifier(eval_metric="TotalF1", depth=9, iterations=580, l2_leaf_reg=12,
                            learning_rate=0.19000000000000003, loss_function='MultiClass',bagging_temperature=0.85,
                            random_strength=120,one_hot_max_size=10,rsm=0.6)


clf.fit(X_train, y_train,cat_features=categorical_features_indices)
```

```python
from sklearn.metrics import f1_score

preds = clf.predict(X_test)
f1_score(y_test, preds, average='micro')
```

```python
#intento con muchos hiperparams
from sklearn.metrics import f1_score

preds = clf.predict(X_test)
f1_score(y_test, preds, average='micro')
```
